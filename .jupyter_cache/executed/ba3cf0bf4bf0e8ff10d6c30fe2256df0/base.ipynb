```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.integrate import odeint
```

```python
def model_malthus(etat, t, params): 
    x = etat        # unpack l'etat
    n, m = params   # unpack params -> params locaux 
    xdot = (n-m)*x  # calcule la derivee de l'etat 
    return xdot     # renvoit la derive 
```

```python
x0 = 0.1

etat0_malthus = np.array([x0])
```

```python
n = 3.0     # taux de natalit
m = 2.0     # taux de mortalit

params_malthus = np.array([n, m])
```

```python
t_0 = 0.0           # temps initial
t_fin = 10.0        # temps final
pas_t = 0.01        # pas de temps 

tspan = np.arange(start=t_0, stop=t_fin, step=pas_t)
```

```python
int_malthus = odeint(
                model_malthus,         # systme dynamique
                etat0_malthus,          # condition initiale
                tspan,                  # temps d'intgration
                args=(params_malthus,), # paramtres du syst. dyn. ici un tuple 
                                        ##  un lment (cf. virgule)
                hmax=pas_t)             # pas d'intgration max. sur temps
```

```python
int_malthus[:5]
```

```python
#| label: fig-malthus
#| fig-cap: simulation et solution mathmatique du modle de Malthus (@eq-malthus)

## figure et systmes d'axes
fig1, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 6))  

## titre de la figure
fig1.suptitle('Simulation du modle de Malthus\n $n = {}, m = {}$'.format(n, m), 
              va='top', fontsize='14')

## premier subplot
ax1.plot(tspan, int_malthus, 
         color='C0', 
         label='$x(t)$ (numrique)')

## modification des bornes
ax1.set_ylim(bottom=None, top=None)

## axes / lgendes / grille
ax1.legend(fontsize='10')
ax1.set_xlabel('Temps $t$', fontsize='12')
ax1.set_ylabel('Densit de population $x$', fontsize='12')
ax1.grid()


## second subplot
ax2.plot(tspan, np.exp((n-m)*tspan)*x0, 
         color='C2', 
         label='$x(t)$ (analytique)')

## axes / lgendes / grille
ax2.legend(fontsize='10')
ax2.set_xlabel('Temps $t$', fontsize='12')
ax2.grid()
```

```python
#| label: fig-erreur-malthus
#| fig-cap: Erreur  l'intgration numrique du modle de Malthus (@eq-malthus)
#| code-fold: true

fig2, ax3 = plt.subplots(1, 1, figsize=(6, 4))
ax3.plot(tspan, 
         int_malthus[:,0]-np.exp((n-m)*tspan)*x0, 
         color='C1', 
         label='$x(t)$ (numrique)')

fig2.suptitle('Erreur d\'intgration pour le modle de Malthus', 
            va='top', fontsize='14')
ax3.set_xlabel('Temps $t$', fontsize='12')
ax3.set_ylabel('Erreur d\'intgration', fontsize='12')
ax3.grid()
```

```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#| code-fold: true

def model_logistic(etat, t, params): 
    x = etat        
    r, K = params   
    xdot = r*x*(1-x/K)  
    return xdot     
```

```python
#| code-fold: true

## condition initiale
x0 = 0.1
etat0_logistic = np.array([x0])

## paramtres
r = 1.0     # taux de croissance intrinsque
K = 10.0    # capacit de charge
params_logistic = np.array([r, K])

## tspan
t_0 = 0.0           
t_fin = 10.0        
pas_t = 0.01        
tspan = np.arange(t_0, t_fin, pas_t)
```

```python
#| code-fold: true

int_logistic = odeint(
                model_logistic,        
                etat0_logistic,         
                tspan,                  
                args=(params_logistic,),
                hmax=pas_t)             
```

```python
#| code-fold: true
#| label: fig-logistic
#| fig-cap: simulation du modle logistique (@eq-logistic)

## figure et systmes d'axes
fig, ax = plt.subplots(1, 1)  

## titre de la figure
fig.suptitle('Simulation du modle logistique ; $r = {}, K = {}$'.format(r, K), 
              va='top', fontsize='14')

## simulation
ax.plot(tspan, int_logistic, 
         color='C0', 
         label='$x(t)$') # solution

## quilibres
ax.plot(tspan, np.zeros_like(tspan),
        color = 'C3',
        linestyle = 'dashed',
        label = "quilibre instable")
ax.plot(tspan, np.ones_like(tspan)*K,
        color = 'C2',
        linestyle = 'dashed',
        label = "quilibre stable")

## modification des bornes
ax.set_ylim(bottom=None, top=None)

## axes / lgendes / grille
ax.legend(fontsize='10')
ax.set_xlabel('Temps $t$', fontsize='12')
ax.set_ylabel('Densit de population $x$', fontsize='12')
ax.grid()
```

```python
#| code-fold: true
%reset -f

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#| code-fold: true

def model_allee(etat, t, params): 
    x = etat        
    r, K, epsilon = params   
    xdot = r*x*(x/epsilon-1)*(1-x/K)  
    return xdot     
```

```python
#| code-fold: true

## paramtres
r = 1.0       # taux de croissance intrinsque
epsilon = 2.0 # seuil de Allee
K = 10.0      # capacit de charge
params_allee = np.array([r, K, epsilon])

## tspan
t_0 = 0.0           
t_fin = 3.0        
pas_t = 0.01        
tspan = np.arange(t_0, t_fin, pas_t)
```

```python
def int_allee(x0, tspan=tspan, params=params_allee):
    sim_allee = odeint(
                model_allee,        
                x0,             # argument de la fonction 
                tspan,          # argument de la fonction 
                args=(params,), # argument de la fonction 
                hmax=pas_t)
    return sim_allee
```

```python
int_allee(0.1)[:5,0]
```

```python
x0_step = 1.35
x0_arr = np.arange(x0_step, K, x0_step)
```

```python
#| label: fig-allee
#| fig-cap: simulation du modle avec effets Allee forts (@eq-allee)
fig, ax = plt.subplots(1, 1)  
fig.suptitle('Simulation du modle avec effets Allee'\
    '; $r={}, K={}, \epsilon={}$'.format(r, K, epsilon), 
              va='top', fontsize='14')

## redfinition du cycle des couleurs pour un dgrad de bleu
colorAllee = plt.cm.Blues(np.linspace(.8, .3, x0_arr.shape[0]))
ax.set_prop_cycle(color = colorAllee)

## simulations
ax.plot(tspan, int_allee(x0_arr[0]),
        label = "$x(t)$")
for x0 in x0_arr[1:]:       # x0 parcour x0_arr
    ax.plot(tspan, int_allee(x0)) 

## quilibres
ax.plot(tspan, np.zeros_like(tspan),
        color = 'C2',
        linestyle = 'dashed',
        label = "quilibre stable")
ax.plot(tspan, np.ones_like(tspan)*K,
        color = 'C2',
        linestyle = 'dashed')
ax.plot(tspan, np.ones_like(tspan)*epsilon,
        color = 'C3',
        linestyle = 'dashed',
        label = "quilibre instable")

## axes / lgendes / grille
ax.legend(fontsize='10')
ax.set_xlabel('Temps $t$', fontsize='12')
ax.set_ylabel('Densit de population $x$', fontsize='12')
ax.grid()
```
